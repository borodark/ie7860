```python
import pandas as pd
import numpy as np
import re
```

```python
fromhdf5 = pd.read_hdf('bdv-data-clean.h5', key='Results')
```

```python
fromhdf5.set_index('BvD ID number',inplace=True)
```

```python
fromhdf5.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
fromhdf5.index
```

```python
yearre= r"[0-9]{4}$" # to catch [2011..2019]
py = re.compile(yearre)
year_dot_re = r"[0-9]{4}\.[0-9]$" # to catch [2011.1..2019.1] 
pyd = re.compile(year_dot_re)
unique_cols = []
strange_ones = []
gifi_columns = []
gifi_dictionary = {}

print(gifi_columns)


only_unique_cols_df =  pd.DataFrame(index=fromhdf5.index)
for col_name in fromhdf5.columns:
    parts = col_name.split()
    year = parts[-1]
    if py.match(year) or pyd.match(year): # one of those <.. 2011> <.. 2011>
        # gifi column
        cleanned_up_col_name = '_'.join(e for e in parts[:-1])
        if not pyd.match(year):
            if cleanned_up_col_name not in gifi_columns:
                gifi_columns.append(cleanned_up_col_name)
            gifi_dictionary[col_name] = (cleanned_up_col_name,int(year))
        print(year + " of " + cleanned_up_col_name)
        if pyd.match(year) and cleanned_up_col_name not in strange_ones:
            strange_ones.append(cleanned_up_col_name)
        
        
    else: # just a column
        cleanned_up_col_name = '_'.join(e for e in parts)
        if cleanned_up_col_name not in unique_cols:
            unique_cols.append(cleanned_up_col_name)
            only_unique_cols_df[cleanned_up_col_name] = fromhdf5[col_name] # 
            print("Added : " + cleanned_up_col_name)
        
```

```python
#gifi_df = None
same_column_df = None
store = pd.HDFStore('bdv-data-gifi.h5')
i=0
last_common_column_name = None
for col_name in fromhdf5.columns:
    if col_name in gifi_dictionary.keys():
        (common_column_name, year) = gifi_dictionary[col_name]
        # if column name changes 
        the_column_of_data = pd.DataFrame(data=fromhdf5[col_name],copy=True)
        print(the_column_of_data.size)
        the_column_of_data.dropna(inplace=True)
        print("After dropna size is:")
        print(the_column_of_data.size)
        the_column_of_data.rename_axis("bv_id",inplace=True)
        the_column_of_data['year'] = year
        the_column_of_data.set_index(['year'], append=True, inplace=True)
        the_column_of_data.rename(columns={col_name: common_column_name},inplace=True)
        # not first column
        if last_common_column_name is None:
            last_common_column_name = common_column_name # deal with the first column
        if not last_common_column_name == common_column_name:
            # new column name 
            store.put(last_common_column_name, same_column_df)
            same_column_df = None
            #if gifi_df is None:
            #    gifi_df = same_column_df
            #else:
                #gifi_df = gifi_df.join(same_column_df, how='outer', on=['bv_id','year'])
                #gifi_df.info(memory_usage=True)
                
            
        # accumulate same column with append
        if same_column_df is not None:
            same_column_df = same_column_df.append(the_column_of_data)
            same_column_df.info(memory_usage=True)
        else:
            same_column_df = the_column_of_data
        print(str(i))
        i=i+1
        last_common_column_name = common_column_name
        #if i > 128:
        #    break

store.close()
```
