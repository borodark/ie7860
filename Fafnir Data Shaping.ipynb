```python
import pandas as pd
import numpy as np
import re
    
```

```python
fromhdf5 = pd.read_hdf('bdv-data-clean.h5', key='Results')
#fromhdf5.replace('n.s.',np.nan, inplace=True)
```

```python
fromhdf5.describe()
```

```python

yearre= r"[0-9]{4}$" # to catch [2011..2019]
py = re.compile(yearre)
year_dot_re = r"[0-9]{4}\.[0-9]$" # to catch [2011.1..2019.1] 
pyd = re.compile(year_dot_re)
unique_cols = []
strange_ones = []
only_unique_cols_df = pd.DataFrame()
for col_name in fromhdf5.columns:
    parts = col_name.split()
    #print(parts)
    year = parts[-1]
    #
    column_data = fromhdf5[col_name]
    if py.match(year) or pyd.match(year): # one of those <.. 2011> <.. 2011>
        cleanned_up_col_name = '_'.join(e for e in parts[:-1])
        print(year + " ++ " + cleanned_up_col_name)
        #TODO add yearly indexed values?
        #column_data[] = 
        if pyd.match(year) and cleanned_up_col_name not in strange_ones:
            strange_ones.append(cleanned_up_col_name)
        if cleanned_up_col_name not in unique_cols:
            unique_cols.append(cleanned_up_col_name)
            # place the col_data into the 'Year->Metric' dataframe
            only_unique_cols_df[cleanned_up_col_name] = column_data 
            # grows with columns
            print("Added : " + cleanned_up_col_name)
        
    else: # just a column
        cleanned_up_col_name = '_'.join(e for e in parts)
        if cleanned_up_col_name not in unique_cols:
            unique_cols.append(cleanned_up_col_name)
            only_unique_cols_df[cleanned_up_col_name] = column_data # 
            print("Added : " + cleanned_up_col_name)
        
#print(unique_cols)
#print(len(unique_cols))
only_unique_cols_df.head()
```
